- Ưu tiên khu vục "region"
- Tìm các post gần nhất bằng Euclidean Distance

```python

df = pd.read_csv("fake-data.csv")
df
```

Step 1: 
    - Tách dataset, ưu tiên khu vực "region"
    - Tạo Minmax Scaler

```python
cal_cols = ["area_by_m2", "width_of_facade", "width_of_road", "is_legal", "price"]
user_last_visited = pd.DataFrame({
    "id": [50],
    "region": ["Di Linh"],
    "area_by_m2": [100],
    "width_of_facade": [10],
    "width_of_road": [5],
    "is_legal": [1],
    "price": [5]
})
df_scaled_data = df.copy()
# df = df.append(user_last_visited.iloc[0])
df = pd.concat([df, user_last_visited])
user_last_visited
df
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(df[cal_cols])

df_scaled_data = df.copy()

df_scaled_data[cal_cols] = scaler.transform(df[cal_cols])
user_last_visited[cal_cols] = scaler.transform(user_last_visited[cal_cols])

df_scaled_data
```

Step 2: 
    - Lọc ra region từ lần cuối truy cập của user
    - Tính Euclid tới các post có cùng khu vực 

```python
last_visited_region = user_last_visited["region"][0]
data_same_region = df_scaled_data[df_scaled_data["region"] == last_visited_region].copy()
data_same_region
```

```python
from scipy.spatial.distance import cdist


data_same_region["score"] = cdist(data_same_region[cal_cols], user_last_visited[cal_cols], metric='euclidean')

df_result = data_same_region.sort_values("score", ascending=True)
recommended_id = df_result["id"].tolist()
print(recommended_id[0:5])
visited_id = user_last_visited["id"].tolist()[0]
recommended_id.remove(visited_id)
df_result
```

Các bài post giống với bài gần nhất user truy cập

```python
recommended_post = pd.DataFrame(columns=df.columns)
for ordered_id in recommended_id:
    recommended_post = pd.concat([recommended_post, df[df["id"] == ordered_id]], ignore_index=True)
recommended_post
```

```python
from Recommender import Recommender
import pandas as pd
import joblib

database = pd.read_csv("fake-data.csv")

scaler = joblib.load("MinMaxScalerModel.gz")
cal_cols = ["area_by_m2", "width_of_facade", "width_of_road", "is_legal", "price"]
based_on = ["id"]

user_last_visited = pd.DataFrame(
    {
        "id": [90],
        "region": ["Di Linh"],
        "area_by_m2": [90],
        "width_of_facade": [5],
        "width_of_road": [5],
        "is_legal": [1],
        "price": [8],
    }
)

recommender = Recommender(
    user_data=user_last_visited,
    database=database,
    cal_cols=cal_cols,
    based_on="region",
    sort_by="id",
    scaler=scaler,
    top=5,
)


print(user_last_visited)
print(recommender.get_recommended())
```

```python
def get_recommended_id(row):
    user_last_visited = pd.DataFrame(
        {
            "id": [row["id"]],
            "region": [row["region"]],
            "area_by_m2": [row["area_by_m2"]],
            "width_of_facade": [row["width_of_facade"]],
            "width_of_road": [row["width_of_road"]],
            "is_legal": [row["is_legal"]],
            "price": [row["price"]],
        }
    )

    recommender = Recommender(
        user_data=user_last_visited,
        database=database,
        cal_cols=cal_cols,
        based_on="region",
        sort_by="id",
        scaler=scaler,
        top=5,
    )

    temp_pd = recommender.get_recommended()
    return temp_pd["id"].to_list()
```

- Ghi lại top-5 giống nhất
- Mỗi lần thêm post mới (hoặc sửa post cũ), sẽ tìm ra 5 post similar mới
- Khi có 5 id mới, duyệt lại 5 id đó, tìm ra 5 id gần nhất trong 6 id (kể cả id của post mới tạo)

Note: giải thích rõ hơn:

1. post mới có `id_mới_nhất` sẽ có 5 id của `5_post_similar`
2. => `5_post_similar` đó sẽ có thay đổi trong cột `top_5_id_similar`
3. => duyệt lại `top_5_id_similar` và thêm `id_mới_nhất` để tìm ra `top_5_id_similar` trong 6 id đó
4. => Mỗi lần thêm/sửa post, sẽ chạy thuật toán tổng 6 lần

Loại tìm bài viết tương tự này có thể duyệt qua profile của user để tìm ra bài viết user có thể có hứng thú.

Mua bán BĐS có tính chất: chỉ bán 1 lần nên ko thể áp dụng lọc cộng tác `collaborative filtering` để gợi ý cho user.

Giả sử ta có user `A`, `B`, `C`.

Case: `A` bán bđs `A'` -> `B` đã mua `A'` -> `A'` không còn trên website

 => ko thể sử dụng lọc cộng tác để gợi ý cho `C` có profile tương tự `B` mua `A'` dc vì `A'` ko còn bán

Bên cạnh đó, nếu sử dụng `collaborative filtering` dựa trên post truy cập của user, ta có thể tìm ra những post dc xem bởi `B`, từ dó recommend cho `C` - người profile tương tự `B`, nhưng sẽ làm loãng những post gợi ý cho `C` dựa theo profile của mình, nếu profile `C` giống hệt `B`, vậy những bài gợi ý cho `B` cũng sẽ được hiển thị cho `C`.
